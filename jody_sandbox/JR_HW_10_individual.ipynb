```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.optimize
sns.set()
```

```python
areas = [460, 455, 460, 455, 470, 465, 464, 469, 471, 473, 475, 477, 483, 483,
         486, 492, 498, 501, 513, 515, 513, 511, 512, 508, 511, 515, 519, 525,
         528, 532, 536, 531, 536, 544, 548, 554, 561, 566, 570, 579, 580, 583,
         582, 586, 582, 588, 584, 591, 599, 596, 600, 607, 609, 613, 622, 630, 638, 
         644, 648, 653, 657, 653, 652, 651, 657, 660, 657, 658,
         663, 667, 659, 659, 670, 677, 690, 701, 704, 711, 722, 723]

times = np.arange(len(areas))*1
```

```python
plt.plot(times,areas, '.')
plt.xlabel("time (min)")
plt.ylabel('bacterial area')
```

```python
fit1 = areas[0]*np.exp(0.10*times)
fit2 = areas[0]*np.exp(0.001*times)
fit3 = areas[0]*np.exp(0.0001*times)

fits = [fit1, fit2, fit3]

for fit in fits:
  plt.plot(times, fit)

plt.plot(times,areas, '.')
plt.xlabel("time (min)")
plt.ylabel('bacterial area')

plt.legend(["fit1","fit2","fit3","areas"])
```
